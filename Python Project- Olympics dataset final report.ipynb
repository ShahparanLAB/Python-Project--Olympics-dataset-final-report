````python
#correlation between GDP and total medals
import pandas as pd

df = pd.DataFrame(olympics_df)
# Compute the correlation between GDP and total medals
correlation = df['gdp'].corr(df['total'])
print(f"The correlation between GDP and the total number of medals won is {correlation:.2f}")
````

````python
#scatter plot with regression line (Question: Is there a correlation between GDP and the total number of medals won by each country?)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.DataFrame(olympics_df)

# Create a scatter plot with regression line
plt.figure(figsize=(12, 8))
sns.regplot(x='gdp', y='total', data=df, scatter_kws={'s':100}, color='blue')
plt.title('Correlation between GDP and Total Medals Won')
plt.xlabel('GDP (in billions)')
plt.ylabel('Total Number of Medals')
plt.show()

````

````python
# 3.Which country has the highest ratio of gold medals to total medals?
````

````python
#highest ratio of gold medals with pandas
import pandas as pd

df = pd.DataFrame(olympics_df)

# Compute the ratio of gold medals to total medals
df['gold_ratio'] = df['gold'] / df['total']

# Find the country with the highest ratio of gold medals to total medals
highest_ratio = df.loc[df['gold_ratio'].idxmax()]

print(f"The country with the highest ratio of gold medals to total medals is {highest_ratio['country']} with a ratio of {highest_ratio['gold_ratio']:.2f}.")
````

````python
#highest ratio of gold medals with seabron
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame(olympics_df)

# Compute the ratio of gold medals to total medals
df['gold_ratio'] = df['gold'] / df['total']

# Plotting the bar chart
plt.figure(figsize=(14, 7))
sns.barplot(x='country', y='gold_ratio', data=df, palette='viridis')
plt.xticks(rotation=90)
plt.title('Ratio of Gold Medals to Total Medals by Country')
plt.xlabel('Country')
plt.ylabel('Gold Medal Ratio')
plt.show()
````

#### Q4: TODO - ask a question here and answer it below

````python
#4.How does the distribution of medals (gold, silver, and bronze) vary across different countries?
````

````python
#distribution of medals (gold, silver, and bronze)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame(olympics_df)

# Melt the DataFrame to long format for easier plotting with seaborn
df_melted = pd.melt(df, id_vars='country', value_vars=['gold', 'silver', 'bronze'], var_name='medal', value_name='count')

# Plot the grouped bar chart
plt.figure(figsize=(18, 8))
sns.barplot(x='country', y='count', hue='medal', data=df_melted, palette='viridis')
plt.xticks(rotation=90)
plt.title('Distribution of Medals (Gold, Silver, Bronze) by Country')
plt.xlabel('Country')
plt.ylabel('Number of Medals')
plt.legend(title='Medal Type')
plt.show()
````

````python
#The country with the smallest gap between gold and silver medals
import pandas as pd
df = pd.DataFrame(olympics_df)

# Compute the gap between gold and silver medals
df['gap'] = abs(df['gold'] - df['silver'])

# Find the countries with the smallest and largest gaps
smallest_gap = df.loc[df['gap'].idxmin()]
largest_gap = df.loc[df['gap'].idxmax()]

print(f"The country with the smallest gap between gold and silver medals is {smallest_gap['country']} with a gap of {smallest_gap['gap']} medals.")
print(f"The country with the largest gap between gold and silver medals is {largest_gap['country']} with a gap of {largest_gap['gap']} medals.")
````
